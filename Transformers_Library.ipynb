####Pipeline :

```python
Pipeline : It Provides abstraction (hide) intermediate steps to make predictions
```

```python
from transformers import pipeline
```

```python
sentiment_pipeline = pipeline(
    task="sentiment-analysis",
    model="siebert/sentiment-roberta-large-english"
)
```

```python
sentiment_pipeline("This is bad product")
```

```python
reviews = [
    "This is good Product",
    "This is bad Product",
    "This is worst Product",
    "This is best Product"
]

sentiment_pipeline(reviews)
```
